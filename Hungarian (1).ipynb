# **Получение матрицы с помощью TfidfVectorizer**

```python
corpus_path = '/content/drive/My Drive/final_300.txt'
token_pattern = '[0-9A-záéúőóüöíÁÉÚŐÓÜÖÍ]+'
with open(corpus_path, 'r') as f:
    vocab = f.readlines()
```

```python
len(vocab)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(token_pattern=token_pattern, min_df=3)
data_vectorized = vectorizer.fit_transform(vocab)

dictionary, idfs = vectorizer.get_feature_names(), vectorizer.idf_
pairs = dict(zip(dictionary, idfs))
```

# **SVD разложение**

```python
from scipy.sparse.linalg import svds
import numpy as np
```

```python
u, sigma, vt = svds(data_vectorized, 10)
vt[:,1]
```

```python
with open('/content/drive/My Drive/final_300_svd_table.npy', 'wb') as f:
        np.save(f, np.dot(np.diag(sigma), vt).T)
```
